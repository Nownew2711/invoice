```python
from pdf2image import convert_from_path
import pytesseract
import pandas as pd
import re

# Convert PDF to images
images = convert_from_path('/home/abinaya/Downloads/test (1).pdf')

# Convert all images to a single text string
full_text = ""
for img in images:
    text = pytesseract.image_to_string(img)
    full_text += text + "\n"  # Add newline to separate pages

# Function to extract required fields from text
def extract_fields(text):
    invoice_no = re.search(r'Invoice No:\s*([A-Z0-9-]+)', text)
    order_no = re.search(r'HM Order No:\s*(\d+)', text)
    invoice_date = re.search(r'Invoice Date:\s*([\d-]+)', text)
    supplier_name = re.search(r'Supplier Name:\s*([A-Z\s]+)', text)
    warehouse_id = None
    warehouse_id_lines = re.findall(r'Warehouse ID\s*', text)

    for match in warehouse_id_lines:
        start = text.find(match)
        next_line_start = text.find('\n', start) + 1
        next_line_end = text.find('\n', next_line_start)
        next_line = text[next_line_start:next_line_end].strip()
        first_word_match = re.match(r'([A-Z0-9]+)', next_line)
        if first_word_match:
            warehouse_id = first_word_match.group(1)
            break
    warehouse_id_match = re.search(r'\bINW\d{3}\b', text)
    warehouse_id = warehouse_id_match.group(0) if warehouse_id_match else None
# Extract description of goods
    description_of_goods = None
    if warehouse_id:
        # Find the line containing the Warehouse ID
        warehouse_line_start = text.find(warehouse_id)
        # Extract text from the same line and subsequent lines until "Container No:" is found
        description_lines = []
        lines = text[warehouse_line_start:].split('\n')
        for line in lines:
            if "Container No:" in line:
                break
            description_lines.append(line.strip())

        # Join all description lines to form the full description
        description = ' '.join(description_lines).replace(' Cartons', '')
        # Split the description into words
        words = description.split()
        # Filter out unwanted words and standalone numbers
        filtered_words = [
            word for word in words
            if not re.match(r'^\d+(\.\d+)?$', word) and word.lower() not in ['usd', 'pieces'] and '=' not in word and not word.startswith('INW')
        ]
        # Rebuild the description
        description_of_goods = ' '.join(filtered_words)

    return {
        'Invoice No': invoice_no.group(1) if invoice_no else None,
        'HM Order No': order_no.group(1) if order_no else None,
        'Invoice Date': invoice_date.group(1) if invoice_date else None,
        'Supplier Name': supplier_name.group(1).strip() if supplier_name else None,
        'Warehouse ID': warehouse_id,
        'Description of Goods': description_of_goods
    }

# Split the full text by occurrences of "Invoice No:" to find each invoice
invoice_texts = full_text.split('INVOICE')

# List to hold all invoice data
data = []

# Process each invoice text
for i in range(1, len(invoice_texts)):
    invoice_text = "INVOICE" + invoice_texts[i]
    fields = extract_fields(invoice_text)
    if fields['Invoice No']:
        data.append(fields)


# Create DataFrame from the data
test_df = pd.DataFrame(data)
# test_df.to_csv("/home/abinaya/Downloads/test_test1.csv")  # Display the DataFrame

# Function to fix hanging words and capitalize text
def fix_description(description):
    words = description.split()
    fixed_words = []
    for i in range(len(words) - 1):
        if len(words[i + 1]) <= 2:  # Check if the next word is a single letter
            fixed_words.append(words[i] + words[i + 1])  # Merge the current word with the next
        else:
            if len(words[i]) > 2:  # Avoid adding single letters alone
                fixed_words.append(words[i])
    # Add the last word if it's not part of a hanging word
    if len(words[-1]) > 1:
        fixed_words.append(words[-1])

    return " ".join(fixed_words).upper()

# Apply the fix to the 'Description of Goods' column
test_df['Description of Goods'] = test_df['Description of Goods'].apply(fix_description)

# Display the updated DataFrame
test_df
test_df.to_csv("/home/abinaya/Downloads/test_test1.csv")  # Display the DataFrame
```

Labels Breakdown:
For each feature, you'll typically have two labels:

B-FeatureName: Beginning of the feature (e.g., "B-INVOICE_NO").
I-FeatureName: Inside the feature (e.g., "I-INVOICE_NO").
Additionally, you'll need an "O" label to indicate tokens that are outside any of the features.

Calculating num_labels:
Invoice No: 2 labels (B-INVOICE_NO, I-INVOICE_NO)
HM Order No: 2 labels (B-HM_ORDER_NO, I-HM_ORDER_NO)
Invoice Date: 2 labels (B-INVOICE_DATE, I-INVOICE_DATE)
Supplier Name: 2 labels (B-SUPPLIER_NAME, I-SUPPLIER_NAME)
Warehouse ID: 2 labels (B-WAREHOUSE_ID, I-WAREHOUSE_ID)
Description of Goods: 2 labels (B-DESCRIPTION_OF_GOODS, I-DESCRIPTION_OF_GOODS)
Outside Label: 1 label ("O")
